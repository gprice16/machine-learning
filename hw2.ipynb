```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


data_path = 'C:\\Users\\gpric\\Downloads/Housing.csv'
df = pd.read_csv(data_path)

print(df.head())

# Compute the cost
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    num_errors = np.subtract(predictions, y)
    sqr_errors = np.square(num_errors)
    cost = np.sum(sqr_errors) / (2*m)
    return cost

# Gradient descent
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha/m) * X.T.dot(errors)
        theta = theta - sum_delta
        cost = compute_cost(X, y, theta)
        cost_history.append(cost)

    return theta, cost_history
# Normalize input
def normalize_input(X):
    minimum_X = X.min(axis=0)
    maximum_X = X.max(axis=0)
    norm_number = np.subtract(X, minimum_X)
    norm_Denom = np.subtract(maximum_X, minimum_X)
    return norm_number / norm_Denom


X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'parking']]
y = df['price']

X_training, X_eval, y_training, y_eval = train_test_split(X, y, test_size=0.2, random_state=50)

X_training = np.array(X_training)
y_training = np.array(y_training)
X_eval = np.array(X_eval)
y_eval = np.array(y_eval)

# Defining the learning rates to explore
learning_rates = [0.01]

# Number of iterations for gradient descent
iterations = 1000

theta = np.zeros(X_training.shape[1])

print("Problem 1a")

# Training the model
plt.figure(figsize=(12, 8))

for lr in learning_rates:
    theta = np.zeros(X_training.shape[1])
    theta, cost_history = gradient_descent(X_training, y_training, theta, lr, iterations)
    theta, cost_history2 = gradient_descent(X_eval, y_eval, theta, lr, iterations)
    plt.plot(range(iterations), cost_history, label=f'Training Losses')
    plt.plot(range(iterations), cost_history2, label=f'Validation Losses')

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost for non-normalized/non-standardized data')
plt.legend()
plt.show()

# Which features impact model the most
plt.figure(figsize=(17, 10))

plt.bar(range(len(theta)), theta) 
plt.xticks(range(len(theta)), X.columns)
plt.xlabel("Features")
plt.ylabel("Coefficients (Weights)")
plt.title("Feature Importance")
plt.show()

X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea']]
y = df['price']

for col in ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']:
    X[col] = X[col].map({'yes': 1, 'no': 0})  # Map 'yes' to 1 and 'no' to 0

X_training, X_eval, y_training, y_eval = train_test_split(X, y, test_size=0.2, random_state=50)

X_training = np.array(X_training)
y_training = np.array(y_training)
X_eval = np.array(X_eval)
y_eval = np.array(y_eval)

# Defining the learning rates to explore
learning_rates = [0.01]

# Number of iterations for gradient descent
iterations = 1000

theta = np.zeros(X_training.shape[1])

print("Problem 1b")

# Training the model and plotting the results
plt.figure(figsize=(12, 8))

for lr in learning_rates:
    theta = np.zeros(X_training.shape[1])
    theta, cost_History = gradient_descent(X_training, y_training, theta, lr, iterations)
    theta, cost_history2 = gradient_descent(X_eval, y_eval, theta, lr, iterations)
    plt.plot(range(iterations), cost_history, label=f'Training Losses')
    plt.plot(range(iterations), cost_history2, label=f'Validation Losses')

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost for non-normalized/non-standardized data')
plt.legend()
plt.show()

# Which features impact model the most
plt.figure(figsize=(17, 10))

plt.bar(range(len(theta)), theta)
plt.xticks(range(len(theta)), X.columns) 
plt.xlabel("Features")
plt.ylabel("Coefficients (Weights)")
plt.title("Feature Importance")
plt.show()


X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'parking']]
y = df['price']

X_training, X_eval, y_training, y_eval = train_test_split(X, y, test_size=0.2, random_state=50)

X_training = np.array(X_training)
y_training = np.array(y_training)
X_eval = np.array(X_eval)
y_eval = np.array(y_eval)

from sklearn.preprocessing import StandardScaler
scaler_X = StandardScaler()
X_training_standardized = scaler_X.fit_transform(X_training)
X_eval_standardized = scaler_X.transform(X_eval)

X_training_normalized = normalize_input(X_training)
X_eval_normalized = normalize_input(X_eval)

# Defining the learning rates to explore
learning_rates = [0.01]

# Number of iterations for gradient descent
iterations = 1000

theta = np.zeros(X_training.shape[1])

print("Problem 2a")

# Training the model with Standardized Data
plt.figure(figsize=(12, 8))

for lr in learning_rates:
    theta = np.zeros(X_training.shape[1])
    theta, cost_history = gradient_descent(X_training_standardized, y_training, theta, lr, iterations)
    theta, cost_history2 = gradient_descent(X_eval_standardized, y_eval, theta, lr, iterations)
    plt.plot(range(iterations), cost_history, label=f'Training Losses')
    plt.plot(range(iterations), cost_history2, label=f'Validation Losses')

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost for Standardized data')
plt.legend()
plt.show()

# Training the model with Normalized Data
plt.figure(figsize=(12, 8))

for lr in learning_rates:
    theta = np.zeros(X_training.shape[1])
    theta, cost_history = gradient_descent(X_training_normalized, y_training, theta, lr, iterations)
    theta, cost_history2 = gradient_descent(X_eval_normalized, y_eval, theta, lr, iterations)
    plt.plot(range(iterations), cost_history, label=f'Training Losses')
    plt.plot(range(iterations), cost_history2, label=f'Validation Losses')

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost for Normalized data')
plt.legend()
plt.show()

# Which features impact model the most
plt.figure(figsize=(17, 10))

plt.bar(range(len(theta)), theta)
plt.xticks(range(len(theta)), X.columns)
plt.xlabel("Features")
plt.ylabel("Coefficients (Weights)")
plt.title("Feature Importance")
plt.show()

X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea']]
y = df['price']

for col in ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']:
    X[col] = X[col].map({'yes': 1, 'no': 0})  # Map 'yes' to 1 and 'no' to 0

X_training, X_eval, y_training, y_eval = train_test_split(X, y, test_size=0.2, random_state=50)

X_training = np.array(X_training)
y_training = np.array(y_training)
X_eval = np.array(X_eval)
y_eval = np.array(y_eval)

from sklearn.preprocessing import StandardScaler
scaler_X = StandardScaler()
X_training_standardized = scaler_X.fit_transform(X_training)
X_eval_standardized = scaler_X.transform(X_eval)

X_training_normalized = normalize_input(X_training)
X_eval_normalized = normalize_input(X_eval)

# Defining the learning rates to explore
learning_rates = [0.01]

# Number of iterations for gradient descent
iterations = 1000

theta = np.zeros(X_training.shape[1])

print("Problem 2b")

# Training the model with Standardized Data
plt.figure(figsize=(12, 8))

for lr in learning_rates:
    theta = np.zeros(X_training.shape[1])
    theta, cost_history = gradient_descent(X_training_standardized, y_training, theta, lr, iterations)
    theta, cost_history2 = gradient_descent(X_eval_standardized, y_eval, theta, lr, iterations)
    plt.plot(range(iterations), cost_history, label=f'Training Losses')
    plt.plot(range(iterations), cost_history2, label=f'Validation Losses')

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost for Standardized data')
plt.legend()
plt.show()

# Training the model with Normalized Data
plt.figure(figsize=(12, 8))

for lr in learning_rates:
    theta = np.zeros(X_training.shape[1])
    theta, cost_history = gradient_descent(X_training_normalized, y_training, theta, lr, iterations)
    theta, cost_history2 = gradient_descent(X_eval_normalized, y_eval, theta, lr, iterations)
    plt.plot(range(iterations), cost_history, label=f'Training Losses')
    plt.plot(range(iterations), cost_history2, label=f'Validation Losses')

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost for Normalized data')
plt.legend()
plt.show()

# Which features impact model the most
plt.figure(figsize=(17, 10))

plt.bar(range(len(theta)), theta) # Change X to a range of numbers
plt.xticks(range(len(theta)), X.columns) # Set x-axis labels to feature names
plt.xlabel("Features")
plt.ylabel("Coefficients (Weights)")
plt.title("Feature Importance")
plt.show()


X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'parking']]
y = df['price']

X_training, X_eval, y_training, y_eval = train_test_split(X, y, test_size=0.2, random_state=50)

X_training = np.array(X_training)
y_training = np.array(y_training)
X_eval = np.array(X_eval)
y_eval = np.array(y_eval)

from sklearn.preprocessing import StandardScaler
scaler_X = StandardScaler()
X_training_standardized = scaler_X.fit_transform(X_training)
X_eval_standardized = scaler_X.transform(X_eval)

X_training_normalized = normalize_input(X_training)
X_eval_normalized = normalize_input(X_eval)

# Defining the learning rates to explore
learning_rates = [0.01]

# Number of iterations for gradient descent
iterations = 1000

theta = np.zeros(X_training.shape[1])

print("Problem 3a")

# Parameter value
Lambda = 0.01

# Function to compute the cost with parameter for loss
def compute_cost_param(X, y, theta, Lambda):
    m = len(y)
    predictions = X.dot(theta)
    num_errors = np.subtract(predictions, y)
    sqr_errors = np.square(num_errors)
    cost = np.sum(sqr_errors) / (2*m) + (Lambda/(2*m)) * np.sum(theta[1:]**2)
    return cost

# Function to perform gradient descent with a parameter
def gradient_descent_param(X, y, theta, alpha, iterations, Lambda):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        predictions = X.dot(theta)
        theta[0] = theta[0] - (alpha/m) * np.sum(predictions - y)
        theta[1:] = theta[1:] - (alpha/m)*X[:, 1:].T.dot(predictions - y) + (Lambda * theta[1:])
        cost = compute_cost_param(X, y, theta, Lambda)
        cost_history.append(cost)

    return theta, cost_history

# Training the model with Standardized Data
plt.figure(figsize=(12, 8))

for lr in learning_rates:
    theta = np.zeros(X_training.shape[1])
    theta, cost_history = gradient_descent_param(X_training_standardized, y_training, theta, lr, iterations, Lambda)
    theta, cost_history2 = gradient_descent_param(X_eval_standardized, y_eval, theta, lr, iterations, Lambda)
    plt.plot(range(iterations), cost_history, label=f'Training Losses')
    plt.plot(range(iterations), cost_history2, label=f'Validation Losses')

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost for Standardized data with parameter')
plt.legend()
plt.show()

# Training the model with Normalized Data
plt.figure(figsize=(12, 8))

for lr in learning_rates:
    theta = np.zeros(X_training.shape[1])
    theta, cost_history = gradient_descent_param(X_training_normalized, y_training, theta, lr, iterations, Lambda)
    theta, cost_history2 = gradient_descent_param(X_eval_normalized, y_eval, theta, lr, iterations, Lambda)
    plt.plot(range(iterations), cost_history, label=f'Training Losses')
    plt.plot(range(iterations), cost_history2, label=f'Validation Losses')

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost for Normalized data with parameter')
plt.legend()
plt.show()

# Which features impact model the most
plt.figure(figsize=(14, 5))

plt.bar(range(len(theta)), theta) 
plt.xticks(range(len(theta)), X.columns) 
plt.xlabel("Features")
plt.ylabel("Coefficients (Weights)")
plt.title("Feature Importance")
plt.show()

X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea']]
y = df['price']

for col in ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']:
    X[col] = X[col].map({'yes': 1, 'no': 0})  # Map 'yes' to 1 and 'no' to 0

X_training, X_eval, y_training, y_eval = train_test_split(X, y, test_size=0.2, random_state=50)

X_training = np.array(X_training)
y_training = np.array(y_training)
X_eval = np.array(X_eval)
y_eval = np.array(y_eval)

from sklearn.preprocessing import StandardScaler
scaler_X = StandardScaler()
X_training_standardized = scaler_X.fit_transform(X_training)
X_eval_standardized = scaler_X.transform(X_eval)

X_training_normalized = normalize_input(X_training)
X_eval_normalized = normalize_input(X_eval)

# Defining the learning rates to explore
learning_rates = [0.01]

# Number of iterations for gradient descent
iterations = 1000

theta = np.zeros(X_training.shape[1])

print("Problem 3b")

# Parameter value
Lambda = 0.01

# Function to compute the cost with parameter for loss
def compute_cost_param(X, y, theta, Lambda):
    m = len(y)
    predictions = X.dot(theta)
    num_errors = np.subtract(predictions, y)
    sqr_errors = np.square(num_errors)
    cost = np.sum(sqr_errors) / (2*m) + (Lambda/(2*m)) * np.sum(theta[1:]**2)
    return cost

# Function to perform gradient descent with a parameter
def gradient_descent_param(X, y, theta, alpha, iterations, Lambda):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        predictions = X.dot(theta)
        theta[0] = theta[0] - (alpha/m) * np.sum(predictions - y)
        theta[1:] = theta[1:] - (alpha/m)*X[:, 1:].T.dot(predictions - y) + (Lambda * theta[1:])
        cost = compute_cost_param(X, y, theta, Lambda)
        cost_history.append(cost)

    return theta, cost_history

# Training the model with Standardized Data
plt.figure(figsize=(12, 5))

for lr in learning_rates:
    theta = np.zeros(X_training.shape[1])
    theta, cost_history = gradient_descent_param(X_training_standardized, y_training, theta, lr, iterations, Lambda)
    theta, cost_history2 = gradient_descent_param(X_eval_standardized, y_eval, theta, lr, iterations, Lambda)
    plt.plot(range(iterations), cost_history, label=f'Training Losses')
    plt.plot(range(iterations), cost_history2, label=f'Validation Losses')

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost for Standardized data with parameter')
plt.legend()
plt.show()

# Training the model with Normalized Data
plt.figure(figsize=(12, 5))

for lr in learning_rates:
    theta = np.zeros(X_training.shape[1])
    theta, cost_history = gradient_descent_param(X_training_normalized, y_training, theta, lr, iterations, Lambda)
    theta, cost_history2 = gradient_descent_param(X_eval_normalized, y_eval, theta, lr, iterations, Lambda)
    plt.plot(range(iterations), cost_history, label=f'Training Losses')
    plt.plot(range(iterations), cost_history2, label=f'Validation Losses')

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost for Normalized data with parameter')
plt.legend()
plt.show()

# Comparing which features impact model the most
plt.figure(figsize=(17, 5))

plt.bar(range(len(theta)), theta)
plt.xticks(range(len(theta)), X.columns)
plt.xlabel("Features") 
plt.ylabel("Coefficients (Weights)")
plt.title("Feature Importance")
plt.show()
```
